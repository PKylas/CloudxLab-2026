```python
#magnify and scale down images: WIP
## Linear interpolation??
import numpy as np
bird = plt.imread('birdie.jpg')

def resized_img(bird, factor):
    h,w,c = bird.shape
    new_h = h//factor
    new_w = w//factor
    res = bird[:new_h*factor, :new_w*factor].reshape(new_h, factor, new_w, factor, c)
    return res.mean(axis=(1, 3)).astype(bird.dtype)
    
downsize = resized_img(bird, 2)    
plt.imshow(downsize)
```
